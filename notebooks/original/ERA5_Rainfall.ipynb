```python
import xarray as xr
import numpy as np

# Correctly extract the total precipitation ('tp') and snowfall ('sf') variables
# Use `.isel` to handle any extra dimensions (if 'expver' is not needed)
total_precip = ds['tp'].isel(expver=0)  # Select the first experiment version if there are multiple
snowfall = ds['sf'].isel(expver=0)

# Select data for January 1979
precip_jan_1979 = total_precip.sel(time='1979-01')
snowfall_jan_1979 = snowfall.sel(time='1979-01')

# Calculate rainfall (total precipitation - snowfall) and convert to mm
rainfall_jan_1979 = (precip_jan_1979 - snowfall_jan_1979) * 1000  # Convert to mm

# Sum over spatial dimensions
total_rainfall_jan_1979 = rainfall_jan_1979.sum(dim=['latitude', 'longitude'])

print(f"Total rainfall for January 1979 (in mm): {total_rainfall_jan_1979.values}")

# To calculate total annual rainfall for 1979, repeat this for each month and sum
rainfall_1979 = 0
for month in range(1, 13):
    # Format the month as two digits
    month_str = f"{month:02d}"
    precip = total_precip.sel(time=f'1979-{month_str}')
    snow = snowfall.sel(time=f'1979-{month_str}')
    rainfall_month = (precip - snow) * 1000  # Convert to mm
    total_rainfall_month = rainfall_month.sum(dim=['latitude', 'longitude'])
    rainfall_1979 += total_rainfall_month

print(f"Total annual rainfall for 1979 (in mm): {rainfall_1979.values}")
```
